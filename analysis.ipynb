```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
import sys
# print(sys.executable)
```

```python
file=pd.read_csv("ecommerce_sales_analysis.csv")
df=pd.DataFrame(file)
print(df)
```

```python
info= df.info()
# print(info)
```

```python
# most_common_product = df["product_name"].value_counts()
# print(most_common_product)


repeated= (df.groupby("category")["product_name"].value_counts().groupby(level=0).head(1))
print(repeated)

u=df['category'].unique()
print(u)
```

```python
monthly_sales_cols = [f"sales_month_{i}" for i in range(1, 13)]
df["total_sales"] = df[monthly_sales_cols].sum(axis=1)
print(df[["total_sales"]])

# Average 
df["average_monthly_sales"] = df["total_sales"] / 12
print(df[["product_id", "product_name", "total_sales", "average_monthly_sales"]])
```

```python
review_score=df.nlargest(5,"review_score")[["product_name","category","review_score"]]
# print(review_score)
review_score_by_category=df.groupby("category")["review_score"].mean()
print(review_score_by_category)

```

```python
# Top 5 products by total sales
top_products = df.nlargest(5, "total_sales")[["product_name", "category", "total_sales"]]
print("Top 5 Products by Total Sales:")
print(top_products)
```

```python
# Total sales per category 
category_sales = df.groupby("category")["total_sales"].sum().sort_values(ascending=False) 
print("\nTotal Sales by Category:") 
print(category_sales)
```

```python
# Average review score per category
category_reviews = df.groupby("category")["review_score"].mean().sort_values(ascending=False)
print("\nAverage Review Score by Category:")
print(category_reviews)
```

```python
# Chart 1 Total sales per category
plt.figure(figsize=(8,5))
category_sales.plot(kind='bar', color='skyblue')
plt.title("Total Sales by Category", fontsize=14)
plt.ylabel("Total Sales", fontsize=12)
plt.xlabel("Category", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Top 5 products by total sales
plt.figure(figsize=(8,5))
plt.bar(top_products["product_name"], top_products["total_sales"], color='orange')
plt.title("Top 5 Products by Total Sales", fontsize=14)
plt.ylabel("Total Sales", fontsize=12)
plt.xlabel("Product Name", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Chart 3: Review score distribution by category
plt.figure(figsize=(8,5))
df.boxplot(column="review_score", by="category", grid=False)
plt.title("Review Score Distribution by Category", fontsize=14)
plt.ylabel("Review Score", fontsize=12)
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Sum sales for each month across all products
monthly_sales = df[[f"sales_month_{i}" for i in range(1, 13)]].sum()

plt.figure(figsize=(8,5))
monthly_sales.plot(kind='bar', color='skyblue')
plt.title("Total Sales by Month", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.xticks(ticks=range(12), labels=[f"Month {i}" for i in range(1, 13)], rotation=45)
plt.tight_layout()
plt.show()
```

```python
# Group by category, then sum monthly sales
category_monthly_sales = df.groupby("category")[[f"sales_month_{i}" for i in range(1, 13)]].sum().T

category_monthly_sales.plot(figsize=(10,6))
plt.title("Monthly Sales by Category", fontsize=14)
plt.xlabel("Month")
plt.ylabel("Sales")
plt.legend(title="Category", bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()
```
